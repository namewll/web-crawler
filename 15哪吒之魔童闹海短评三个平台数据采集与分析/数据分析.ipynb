```python
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams["font.family"]=["Microsoft Yahei","SimHei","sans-serif"]
```

```python
df1=pd.read_excel("哪吒之魔童闹海短评.xlsx","Sheet1")
df2=pd.read_excel("nezha.xlsx","Sheet1")
df3=pd.read_excel("imdb.xlsx","Sheet1")
df3
```

```python
star_counts=df1["星级"].value_counts()
plt.figure(figsize=(5,7))
plt.pie(star_counts,autopct="%1.1f%%",labels=star_counts.index,explode=(0.1,0,0,0,0,0))
plt.title("哪吒之魔童闹海在豆瓣网站的用户评价星级分布")
plt.show()
```

```python
star_counts=df2["星级"].value_counts()
plt.figure(figsize=(10,14))
plt.pie(star_counts,autopct="%1.1f%%",labels=star_counts.index,explode=(0.2,0,0,0,0,0,0,0,0,0),pctdistance=0.75,labeldistance=1.15,textprops={"fontsize": 9})
plt.title("哪吒之魔童闹海在Rotten Tomatoes网站的用户评价星级分布")
plt.show()
```

```python
star_counts=df3["星级"].value_counts()
plt.figure(figsize=(5,7))
plt.pie(star_counts,autopct="%1.1f%%",labels=star_counts.index,explode=(0.1,0,0,0))
plt.title("哪吒之魔童闹海在IMDb网站的用户评价星级分布")
plt.show()
```
